```python
def pick_random_smooth_integer(S,num_factors):
    return math.prod(random.sample(S,num_factors))**2

def bits(x):
    ''' returns the number of bits of the positive integer x'''
    if x>0:
        return math.floor(math.log(x,2))+1;
    else:
        return 0


def decimalToBinary(n):
    return bin(n).replace("0b", "")

def string_to_list_of_ints(l):
    l = list(l)
    for i in range(len(l)):
        l[i] = int(l[i])  # make elements int
    return l


def random_prime(bits):
    while 1==1:
        N = random.getrandbits(bits)
        if millerRabin(N, 8):
            break
    return N

def rsa_modulus(bits):
    #creating random prime numbers
    p = random_prime(int(np.floor(bits/2)))
    q = random_prime(int(np.floor(bits/2)))
    #checking if p and q is equal, until I create different prime numbers
    if p!=q:
        return p,q,p*q

def legendreSymbol(a, p):
    return pow(a, (p - 1) // 2, p)

def pi(x):
    return primesieve.count_primes(x)

def legendre_symbol(n, k):
    # SourceL: https://en.wikipedia.org/wiki/Jacobi_symbol
    if n==1:return 1
    assert(k > 0 and k % 2 == 1)
    n = n % k
    t = 1
    while n != 0:
        while n % 2 == 0:
            n = n / 2
            r = k % 8
            if (r == 3 or r == 5):
                t = -t
        n, k = k, n
        if (n % 4 == 3 and k % 4 == 3):
            t = -t
        n = n % k
    if k == 1:
        return t
    else:
        return 0
    
def Ln(x):
    # Function to calculate the B-limit
    return math.exp(math.sqrt(math.log(x)*(math.log(math.log(x)))))

def is_square(n):
    return math.sqrt(n).is_integer()
```

```python
def fast(base, power, modulo):
    g = decimalToBinary(power)
    g = string_to_list_of_ints(g)
    x = base
    d = 1
    for i in range(len(g) - 1, -1, -1):
        if g[i] == 1:
            d = d * x % modulo
        x = (x * x) % modulo
    return d

def millerRabin(n, k):
    import random
    i = 1
    j = 0
    if n & 1 == 0:    #print C
    #if R=="todo":
    #    return "todo"
        return False
    else:
        """
        write n-1 as 2^s * t found from
        https://en.wikibooks.org/wiki/Algorithm_Implementation/Mathematics/Primality_Testing
        fbclid=IwAR3AENaDRikz49KC5TiZngp4nOda19z3c8tqpisZop7A1Ju0ZdIoS7WoZrE
        """
        s, t = 0, n - 1
        while t & 1 == 0:
            s, t = s + 1, t >> 1
        while i <= k:
            a = random.randint(2, n - 1)
            b = fast(a, t, n)
            if b % n == 1:
                j = j + 1
                if j == k:
                    return True
            for r in range(s):
                if b % n == n - 1:
                    j = j + 1
                    if j == k:
                        return True
                b = fast(b, 2, n)
            i = i + 1
        return False


def isPrime(n):
    import primesieve
    lowPrimes = primesieve.n_primes(100)
    if n >= 3:
        if n & 1 != 0:
            if n in lowPrimes:
                return True
            return millerRabin(n, 2)
    return False
```

```python
# calculates x so that x ^ 2 = a mod p
# with Tonelli Shanks Algorithm
# parameters: p prime
#             a integer
# return : x
def tonelliShanks (a, p):
    if a==1:return  1,-1%p
    if is_square(a):return int(math.sqrt(a)),-int(math.sqrt(a))%p
    if isPrime(p):
        
        # check that a is a square: (n | p) = 1
        if (legendreSymbol(a,p) == 1):


            # Partition p-1 to q * 2^e for an odd q 
            # (reduce all the powers of 2 from p - 1)
            q = p - 1
            e = 0
            while q % 2 == 0:
                q = q // 2
                e += 1

            # find z so that z|p = -1
            z = 2
            while (p - 1 != legendreSymbol(z, p)):
                z += 1

            c = pow(z, q, p)
            r = pow(a, (q + 1) // 2, p)
            t = pow(a, q, p)
            m = e
            v = 0

            # loop until t = 1
            while (t - 1) % p != 0:
                v = (t * t) % p

                for i in range(1, m):
                    if (v - 1) % p == 0:
                        break

                    v = (v * v) % p

                b = pow(c, pow(2, m - i - 1), p)
                r = (r * b) % p
                c = (b * b) % p
                t = (t * c) % p
                m = i

            return r,-r%p
        else:
            return "a,p, there is not any solution :",a,p
    else:
        return "p is not prime"


```

```python
# sieve for B-smooth numbers
# on input J>I>0, a finite set of primes P and a RSA modulus N
# this algorithm finds the factors in set P of numbers x^2-N as x runs in {I,...,J}
  
import itertools
import operator
def wagstaff_algorithm(I,J,P,N):
    
    def least(I,J,q,root): # least integer >=I such that =root mod q
        for i in range(I,J):
            if i%q==root:
                return i
            
    # printing elegant
    def multiplicity(n,prime_divisor):
        i = 1
        while prime_divisor<n and n%prime_divisor==0:
            n = n // prime_divisor
            i +=1
        return i-1
    
    L = []
    for q in P:
        x = tonelliShanks(N,q) # compute the two roots x^2=N mod q
        if x!=None:
            if x[0]!=x[1]: # the roots x[0],x[1] perhaps are equal.
                for j in range(2):
                    i = least(I,J,q,x[j])
                    if i!=None:
                        while i<J:
                            L.append([i,q])
                            i = i + q
            else:
                i = least(I,J,q,x[0]) # we have one root with multiplicity 2
                if i!=None:
                    while i<J:
                        L.append([i,q])
                        i = i + q
    V=sorted(L, key=lambda x: x[0]) # we sort the result
    print("Wagstaff's algorithm found:",len(V), "numbers")
    W = []
    key_func = operator.itemgetter(0)
    for key, group in itertools.groupby(V, key_func):
        W.append([key] + [i[1:] for i in list(group)])
    return W


# now we shall find B-smooth numbers 
# of the form x^2-N for x running to an interval [I,J] 
# using the previous algorithm

def Bsmooth_qs(I,J,N,B,num_of_relations):
    # I lower bound
    # J upper bound
    # N the rsa modulus
    # the bound for the smooth numbers, common value is ~sqrt(L(N))
    # when you find num_of_relations solutions stop the algorithm
    
    from math import sqrt
    
    def is_square(n):
        return sqrt(n).is_integer()
    
    def multiplicity(n,prime_divisor):
        i = 1
        while prime_divisor<n and n%prime_divisor==0:
            n = n // prime_divisor
            i +=1
        return i-1
    
    smooth=[]
    P = primesieve.primes(B) 
    # for QS we further demand the factor base P to contain 
    # primes of the form legendre(N,p)=1
    # so we have to remove all the primes p of P such that legendre(N,p)=-1
    P_QS = [2]+[p  for p in P[1::] if legendreSymbol(N,p)==1]
    W = wagstaff_algorithm(I,J,P_QS,N)
    for j in range(len(W)):
        a = W[j]
        local_product = math.prod([a[i][0]**multiplicity(a[0]**2-N,a[i][0]) for i in range(1,len(a))])
        if local_product == a[0]**2 - N:
            # make a further check if a[0]^2-N is a square
            # in this case, and if you are lucky, you can find
            # a factor of N
            if is_square(local_product):
                sqr = int(math.sqrt(local_product))
                if math.gcd(a[0]-sqr,N)>1:
                    print("Lucky! :",math.gcd(a[0]-sqr,N))
                    return 
            
            # add the multiplicity to each factor
            for i in range(1,len(a)):
                a[i].append(multiplicity(a[0]**2-N,a[i][0]))
                
            
            smooth.append(a)
            if len(smooth)==num_of_relations:
                break
            
    print("B-smooth numbers found",len(smooth))
    return smooth
```

```python
def trial_division(n):
    import gmpy2 as gmp
    counter   = 0
    counter_2 = 0
    a = []
    if (n<0): # if n is negative add -1 to list
        a.append(-1)
        n=n*(-1)
    while n % 2 == 0:
        a.append(2)
        n = gmp.c_div(n, 2)
        counter = counter + 1
        counter_2 = counter_2 + counter
    f = 3 # f is a possible prime divisor. We shall search up to sqrt(n).
    while gmp.mpz(f)**2<= n:
        if n % f == 0: # similar as previous. We check if f divides n.
            a.append(f) 
            #n /= f    # we update n with n/f
            n = gmp.c_div(n, f)
            counter = counter + 1
            counter_2 = counter_2 + counter
        else:
            f += 2   # we increase the possible divisor f by 2. So f = 3,5,7,9,11....
            counter_2 = counter_2 +  1 
    if n != 1: a.append(gmp.mpz(n)) # put n in the list of divisors of n
    # Only odd number is possible
    return a

def is_smooth(b,S):
    # check if integer b has divisors in S
    b_fact=[i for i in trial_division(b)]
    for i in range(len(b_fact)):
        if not b_fact[i] in S:
            return False
    return True
```

```python
def row_echelon(A): 
    """
    https://math.stackexchange.com/questions/3073083/how-to-reduce-matrix-into-row-echelon-form-in-python
    """
    """ Return Row Echelon Form of matrix A """

    # if matrix A has no columns or rows,
    # it is already in REF, so we return itself
    r, c = A.shape
    if r == 0 or c == 0:
        return A

    # we search for non-zero element in the first column
    for i in range(len(A)):
        if A[i,0] != 0:
            break
    else:
        # if all elements in the first column is zero,
        # we perform REF on matrix from second column
        B = row_echelon(A[:,1:])
        # and then add the first zero-column back
        return np.hstack([A[:,:1], B])

    # if non-zero element happens not in the first row,
    # we switch rows
    if i > 0:
        ith_row = A[i].copy()
        A[i] = A[0]
        A[0] = ith_row

    # we divide first row by first element in it
    A[0] = A[0] / A[0,0]
    # we subtract all subsequent rows with first row (it has 1 now as first element)
    # multiplied by the corresponding element in the first column
    A[1:] -= A[0] * A[1:,0:1]

    # we perform REF on matrix from second row, from second column
    B = row_echelon(A[1:,1:])

    # we add first row and first (zero) column, and return
    return np.vstack([A[:1], np.hstack([A[1:,:1], B]) ])
```

```python
def factor_(B,A,sol,num):
    Z = 1
    X2 = 1
    for i in range(len(sol)):
        if sol[i]==1:
            Z = Z*B[i]
            X2 = X2*A[i]**2
    return gcd(sqrt(Z)-sqrt(X2),num)

def pivot(v):
    # v is a binary list
    # output the entry of the first 1
    for i in range(len(v)):
        if v[i]==1:
            return i
        
def are_the_same(list1, list2):
    if len(list1)!=len(list2):
        return False
    else:
        for i in range(len(list1)):
            if list1[i]!=list2[i]:
                return False
        return True 
    
def take_cols(M):   
    # it takes a matrix M and return a list with the non zero columns of M
    num_rows, num_cols = M.shape
    dim=(num_rows,num_cols)
    col = []
    col.append(  [ M[0:dim[0],i] for i in range(dim[1]) ] )
    col = col[0]

    C = []
    for i in range(len(col)):
        if not(all(v == 0 for v in col[i])):
            C.append(col[i])
    return C

def initialization_for_C(M):
    C = take_cols(M)
    
    temp1 = (len(C[0])-1)*[0]+[1]
    temp2 = C[len(C)-1]       
    if are_the_same(temp1,temp2):
        C.pop(len(C)-1)
        return C,1
    else:
        return C,-1000
    
def take_the_free_unknowns(C):
    # all the elements under the pivot of last element of C corresponds to free unknowns
    lst = []
    lenC = len(C)
    #l = matrix(C).dimensions()[1]
    num_rows, num_cols = np.array(C).shape
    l=num_cols
    piv = pivot(C[lenC-1])
    temp = piv + 1
    L = np.array((np.array(C).transpose()[temp:l]).transpose())
    lst = [i for i in range(piv+1,l)]       
    for i in range(lenC-2,-1,-1):
        a = pivot(C[i+1])-pivot(C[i])
        if a>1:
            lst1 = []
            lst1 =  [j for j in range(pivot(C[i])+1,pivot(C[i])+a)]
            lst = lst + lst1
            R1 = np.array( (np.array(C).transpose()[pivot(C[i])+1:pivot(C[i])+a]).transpose() )
            L = np.block([[R1,L]])
    lst = flatten(lst)[::-1]

    num_rows, num_cols = L.shape

    return L,num_cols,sorted(flatten(lst))

def pick_a_random_solution(T):
    C = T[0]
    R = take_the_free_unknowns(C)
    v= (np.random.randint(2,size=(1,R[1]))).transpose()
    basic_unknowns = np.matmul(R[0],v)
    lst = [-1 for i in range(len(C[0]))]
    
    j = 0
    for i in R[2]:
        lst[i] = v[j][0]
        j = j + 1
    j = 0
    for i in range(len(C[0])): 
        if int(lst[i])==-1:
            lst[i] = basic_unknowns[j][0]
            j = j + 1
    if T[1]==-1000:
        lst = lst + [0]
    return lst
```

```python
def exponent(set, base):
    # returns how many times "base" exists in "set"
    count = 0 
    for i in range(len(set)):
        if set[i]==base:
            count+=1
    return count
```
